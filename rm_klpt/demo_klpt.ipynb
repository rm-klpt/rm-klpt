```python
d = 5
p = (2^128).next_prime()
while p % 4 != 3:
    p = p.next_prime()
right = initial_rm(d, p)
```

```python
##We take a short walk in the RM 2-isogeny graph to generate a small randomish RM-pair
left = right.random_walk(10)
gamma_128_short = right.klpt(left, check=True)
```

```python
#The output isogeny we obtain here has larger degree than optimal. This is to be expected: since we took a short random walk, the lattice
#RM-homomorphisms from left to right has an element of degree 2^10, and is therefore unbalanced. This tends to increase the degree of the
#equivalent prime ideal that we find during the klpt computation. In turn, this affects the quality of our output.
```

```python
#Because of coefficient explosion, running the computation with the output of a long random walk is somewhat costly.
#We first to the experiment with a smaller prime. Here, with a 32 bits p, we may afford a long, well-mixing walk.
```

```python
d = 5
p = (2^32).next_prime()
while p % 4 != 3:
    p = p.next_prime()
right = initial_rm(d, p)
left = right.random_walk(40)
gamma_32_long = right.klpt(left, check=True)
```

```python
#We propose finally a well-mixed instance with a 128 bits prime. Please take into consideration that the computation below
#takes over 30 minutes on a 5GHz AMD Ryzen 5 7600 CPU
```

```python
d = 5
p = (2^128).next_prime()
while p % 4 != 3:
    p = p.next_prime()
right = initial_rm(d, p)
left = right.random_walk(70)
gamma_128_long = right.klpt(left, check=True)
```
